```python
import re
def removepunct(text):
  joined_filtered_sentence = re.sub(r'[^a-zA-Z0-9 ]',r'',text)
  return joined_filtered_sentence
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
```

```python

def removestopword(text): 
  sentence = nlp(text)
  notStopWords = [notStopWords.text for notStopWords in sentence if not notStopWords.is_stop]  
  joined = " ".join(notStopWords)
  return joined
```

```python
# df['Heading']=df['Heading'].apply(removestopword)
df=df.astype(str)
df=df.applymap(removestopword)
df
```

```python
def find_ORG_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'ORG']
  result=" ".join(entities)
  return result

def find_PERSON_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'PERSON']
  result=" ".join(entities)
  return result 
def find_DATE_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'DATE']
  result=" ".join(entities)
  return result
def find_GPE_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if i.label_ == 'GPE']
  result=" ".join(entities)
  return result
def find_PRODUCT_Ents(text):
  my_doc = nlp(text)
  entities=[i.text for i in my_doc.ents if (i.label_ == 'PRODUCT') ]
  result=" ".join(entities)
  return result    
  
```

```python
df['ORG'] = df['Heading'].apply(find_ORG_Ents)+' '+df['Content'].apply(find_ORG_Ents)
df['PERSON']=df['Heading'].apply(find_PERSON_Ents)+' '+df['Content'].apply(find_PERSON_Ents)
df['GPE']=df['Heading'].apply(find_GPE_Ents)+' '+df['Content'].apply(find_GPE_Ents)
df['PRODUCT']=df['Heading'].apply(find_PRODUCT_Ents)+' '+df['Content'].apply(find_PRODUCT_Ents)
df['DATE']=df['Heading'].apply(find_DATE_Ents)+' '+df['Content'].apply(find_DATE_Ents)
df.head()
```

# **Similarity Search by IVFPQ + HNSW**

```python
!pip install transformers
```

```python
from transformers import AutoTokenizer, TFAutoModel
```

```python
model_ckpt = "sentence-transformers/multi-qa-mpnet-base-dot-v1"
tokenizer = AutoTokenizer.from_pretrained(model_ckpt)
model = TFAutoModel.from_pretrained(model_ckpt, from_pt=True)
```

```python
!pip install datasets
```

```python
from datasets import Dataset

file_dataset = Dataset.from_pandas(df)
file_dataset
```

```python
def concatenate_text(examples):
    return {
        "text": examples["Heading"]
        }


file_dataset = file_dataset.map(concatenate_text)
file_dataset
```

```python
def cls_pooling(model_output):
    return model_output.last_hidden_state[:, 0]
```

```python
def get_embeddings(text_list):
    encoded_input = tokenizer(
        text_list, padding=True, truncation=True, return_tensors="tf"
    )
    encoded_input = {k: v for k, v in encoded_input.items()}
    model_output = model(**encoded_input)
    return cls_pooling(model_output)
```

```python
embedding = get_embeddings(file_dataset["text"])
embedding.shape
embedding.numpy()
```

```python
embeddings_dataset = file_dataset.map(
    lambda x: {"embeddings": get_embeddings(x["text"]).numpy()[0]}
)
   
```

```python
embeddings_dataset.add_faiss_index(column='embeddings')
```

```python
question = input("Search:")
question_embedding = get_embeddings([question]).numpy()
question_embedding
```

```python
scores, samples = embeddings_dataset.get_nearest_examples(
    "embeddings", question_embedding, k=5
)
   
```

```python
samples_df = pd.DataFrame.from_dict(samples)
samples_df["scores"] = scores
samples_df.sort_values("scores", ascending=False, inplace=True)
```

```python
for _, data in samples_df.iterrows():
    print("~" * 200)
    print(f"Titles: {data.text}\nArticle conetent: {data.Content}\nArticle Date: {data.Date}\nArticle Website: {data.Website}")
    print("~" * 200)
    
```

# **Implementation with Faiss:(IndexIVFPQ + HNSW) on mongoDB data.**

```python
xb=embedding.numpy()[:700]
xq=question_embedding
print(xb.shape)
xq.shape
```

```python
import faiss
d = xb.shape[1]        
M = 64        
nlist = xb.shape[0] 
nsegment = 64  
nbit = 8     
k=50

coarse_quantizer = faiss.IndexHNSWFlat(d, M)
index = faiss.IndexIVFPQ(coarse_quantizer, d, nlist, nsegment, nbit)      


```

```python
index.train(xb)
```

```python
index.add(xb)
index.nprobe = 30 
```

```python
Dist, I = index.search(xq, k)
```
